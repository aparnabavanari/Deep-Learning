```python
import csv

#1. function to prompt & Validate the inputs from the user for expense details
def expenseStore():
    
    dailyExpenses=[]
    while True:
        #Prompt to enter the expense date
        expenseDate = input("Enter Date(YYYY-MM-DD) of Expense or 'exit' to stop:").strip()
        if expenseDate.lower() == 'exit':
            break        
            
        #Validate the input date
        if not expenseDate or len(expenseDate)!=10 or not expenseDate.replace("-","").isdigit():
            print("Invalid Date Format. Please Enter a Valid Date")
            continue
        
        #Prompt to enter the expense Category
        expenseCat=input(f"Enter Category of Expense for {expenseDate} :").strip()
        
        #Validate the input category
        if not expenseCat:
            print("Categoty is Mandatory. Please Enter a valid Expense Category")
            continue
        
        #Prompt to enter expense Description details
        expenseDesc=input("Enter Expense Description:").strip()
        
        #validate the Expense description entere
        if not expenseDesc or len(expenseDesc)==0:
            print("Description is Mandatory. Please Enter a valid expense descriptiopn")
            continue
        
        #Prompt to input the Expense Amoutn
        expenseAmt=float(input("Enter Amount of Expense:"))
        
        #Validate the entered Expense Amount
        if expenseAmt<=0:
            print("Amount must be greater than 0. Please enter a valid numeric value for Amount.")            
        
        #store the expense in Dictionary
        expense={
            'Date':expenseDate,
            'Category':expenseCat,
            'Amount':expenseAmt,
            'Description':expenseDesc            
        }
        dailyExpenses.append(expense)
        print(f"Expense Added : {expense}")
    
    return dailyExpenses

#Create a function that allows the user to input a monthly budget. 
#Prompt the user to Enter the total amount they want to budget for the month
def monthlyBdgt():
    monthBdgt=float(input('Enter Monthly Budget'))
    return monthBdgt

#Create another function that calculates the total expenses recorded so far
def totalExpenses(expense_Lst,monthly_Budget):
    total_exp=0.0
    for expense in expense_Lst:
        total_exp=total_exp+expense['Amount']
    print(total_exp)    
    
    #If the total expenses exceed the budget, display a warning (Example: You have exceeded your budget!)
    if(float(total_exp)>float(monthly_Budget)):
        print('You have exceeded your budget!')
    elif(float(total_exp)==monthly_Budget):
        print('Your Budget is sufficient for total expenses')
    else:
        #Compare the total with the user’s monthly budget
        total_differ=monthly_Budget-total_exp
        print(f'You have {total_differ} left for the month')                

#function to save all expenses to a CSV file,
def writeExptoExcel(expenses, filename):
    try:
        with open(filename,mode="w",newline="") as csvfile:
            writer=csv.writer(csvfile)
            #Header Row
            writer.writerow(["Date","Category","Amount","Description"])
            #Write each expense in a row
            for expense in expenses:                
                writer.writerow([expense["Date"],expense["Category"],expense["Amount"],expense["Description"]])
        print(f'Expense Saved to file {filename}')
    except Exception as ex:
        print(f'Error Saving Expenses to .csv file:{ex}')

#function to load expenses from the CSV file.    
def readExpfromExcel(filename):
    expenses=[]
    try:
        with open(filename, mode="r") as csvfile:
            #Read the saved data from the file
            reader=csv.DictReader(csvfile)
            for row in reader:
                expense.append(row)
        print(f'Expenses loaded from csv file: {filename}')
    except FileNotFoundError:
        print(f'{filename} not found.')
    except Exception as ex:
         print(f'Error loading Expenses from .csv file:{ex}')
       
    
#Main Logic where the program starts
expense_Lst=expenseStore()
#print(expense_Lst)
print("Daily Expenses")

#2. Retrieve and display the list of expenses
for expense in expense_Lst:
    print(expense)
    
#3. Set and track the budget:
monthly_Budget = monthlyBdgt()
#print(monthly_Budget)
totalExpenses(expense_Lst,monthly_Budget)

#4.Save\write and load\read expenses from .csv files:
filename="Monthly_Exp_June2025.csv"
writeExptoExcel(expense_Lst, filename)
readExpfromExcel(filename)

#5. Create an interactive menu:

        
```
